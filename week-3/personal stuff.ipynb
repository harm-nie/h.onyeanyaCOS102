```python
from tkinter import *
count = 0

def click():
    global count
    count+=1
    label1.config(text=count)
    print(count)

def submit():
    username = entry.get()
    print(f"Hello {username}")

def backspace():
    entry.delete(len(entry.get())-1,END)

def delete():
    entry.delete(0,END)
window = Tk()

window.title("FIRST GUI")

window.config(background="#e0f0e3")
label = Label(window,text="User's Name",font=('Arial',30,'italic'),fg='#B069DB',bg='#e0f0e3')
label.place(x=100,y=100)

button = Button(window,text='Click me')
button.config(command=click)
button.config(font=('Ink Free',50,'bold'))
button.config(bg='#B069DB', fg='#FFFFFF', activebackground='#FF0000', activeforeground='#FFFFFF')


button.pack()

label = Label(window,text="User's Department",font=('Arial',30,'italic'),fg='#B069DB',bg='#e0f0e3')
label.place(x=100,y=400)

label1 = Label(window,text=count)
label1.config(font=('Monospace', 50))
label1.pack()


entry = Entry()
entry.config(font=('Arial',30,'bold'),fg='#B069DB',bg='#FFFFFF')
#entry.insert(0,'Spongebob')
entry.config(width=10)
entry.pack()

submit = Button(window,text="submit",command=submit )
submit.pack(side = RIGHT)

delete = Button(window,text="clear",command=delete)
delete.pack(side=RIGHT)
backspace = Button(window,text="backspace",command= backspace)
backspace.pack(side = RIGHT)
window.mainloop()

```
