```python
from tkinter import *
import csv

def welcome(name):
    well=Tk()
    well.title("Access granted")
    well.geometry("400x500")
    well.config(bg='#111111')
    label= Label(well, text=f"Welcome\n{name}", font=("Segoe UI", 14, "bold"), fg="#FFFFFF", bg="#111111")
    label.pack(pady=30)


def reject(name):
    well=Tk()
    well.title("Access DENIED")
    well.geometry("400x500")
    well.config(bg='#111111')
    label= Label(well, text=f"Sorry\n{name},\n Access denied" ,font=("Segoe UI", 14, "bold"), fg="#FFFFFF", bg="#111111")
    label.pack(pady=30)
    

def submit():
    name = name_entry.get().title()
    department = dept_entry.get().title()

    found=False
    with open ("C:\\Users\\Onyeanya Harmonie\\h.onyeanyaCOS102\\week-3\\GIG-logistics.csv","r") as csv_file:
        staff_file=csv.DictReader(csv_file)
        for row in staff_file:
            if row['Name'] == name and row['Department'] == department:
                found = True
                break
    if found:
        welcome(name)
    else:
        reject(name)

    print(f"Hello {name}. you're in {department}")


def clear():
    name_entry.delete(0,END)
    dept_entry.delete(0,END)



window = Tk()
window.title("Employee Login")
window.geometry("500x600")
window.config(bg='#111111')

logo = PhotoImage(file="C:\\Users\\Onyeanya Harmonie\\h.onyeanyaCOS102\\week-3\\gig logistics.png")
lbl = Label(image=logo).pack()
logo.config()


#labels and entries 
name_lbl = Label(window, text= "Username", font=('Arial',30,'bold'), fg='#FFFFFF',bg='#111111')
name_lbl.place(x=50,y=180)

name_entry = Entry()
name_entry.config(font=('Arial',20,'bold'),fg='#B069DB',bg='#FFFFFF')
name_entry.place(x=50, y=240, width=300, height=50)


dept_lbl = Label(window, text= "Department", font=('Arial',30,'bold'), fg='#FFFFFF',bg='#111111')
dept_lbl.place(x=50,y=360)

dept_entry = Entry()
dept_entry.config(font=('Arial',20,'bold'),fg='#B069DB',bg='#FFFFFF')
dept_entry.config(width=10)
dept_entry.place(x=50, y=420, width=300, height=50)

submit = Button(window,text="Login",command=submit )
submit.place(x=600, y=520)

delete = Button(window,text="clear",command=clear)
delete.place(x=600, y=600)

window.mainloop()
```
