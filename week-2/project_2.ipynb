```python
import numpy as np 
choice = ""

# the formulas
print("\n--- Equation Root Calculator ---")
print()
print("1. Quadratic Equation: Ax^2 + Bx + C = 0")
print("2. Cubic Equation: Ax^3 + Bx^2 + Cx + D = 0")
print("3. Quartic Equation: Ax^4 + Bx^3 + Cx^2 + Dx + E = 0")

running = True
while running: 
    while choice not in ['1', '2', '3']:

    
        choice = input("Select an option (1, 2, or 3): ")

    # for invalid input
        if choice not in ['1', '2', '3']:
            print()
            print("Invalid input! Please enter only 1, 2, or 3.")

    
    if choice == '1':
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        print()
        coefficients = [a, b, c]
        print(f"The coefficients are: {coefficients}")
        roots = np.roots(coefficients)
        print(f"The roots are: {roots}")

    elif choice == '2':
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        d = float(input("Enter D: "))
        print()
        coefficients = [a, b, c, d]
        print(f"The coefficients are: {coefficients}")
        roots = np.roots(coefficients)
        print(f"The roots are: {roots}")

    elif choice == '3':
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        d = float(input("Enter D: "))
        e = float(input("Enter E: "))
        print()
        coefficients = [a, b, c, d, e]
        print(f"The coefficients are: {coefficients}")
        roots = np.roots(coefficients)
        print(f"The roots are: {roots}")

    print("\n----------------------------------------")
    choice = input("Would you like to run another calculation? (y/n): ").lower()
    
    if choice != 'y':
        break
        print("Exiting program. Goodbye!")

```
