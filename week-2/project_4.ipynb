```python
print("------ANNUAL TAX REVENUE ALLOCATOR------")

running = True

while running:
    year = int(input("Enter your years of experience"))
    age = int(input("Enter your age"))
    print()
    print(f"Years of experience: {year}")
    print(f"Your age: {age}")

    #conditions
    if year > 25 and age >= 55:
        print("Annual Tax Revenue is: 5_600_000 ")

    elif year > 20 and age >= 45:
        print("Annual Tax Revenue is: 4_480_000 ")

    elif year > 10 and age >= 35:
        print("Annual Tax Revenue is: 1_500_000 ")

    elif year < 10 and age <35 :
        print("Annual Tax Revenue is: 550_000 ")

    else:
        print("No ATR allocated")

    print("\n----------------------------------------")
    choice = input("Would you like to run another calculation? (y/n): ").lower()
    
    if choice != 'y':
        running = False
        print("Exiting program. Goodbye!")
```
