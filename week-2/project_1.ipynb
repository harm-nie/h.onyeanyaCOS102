```python
# Part C: Age Swapping Algorithm

name1 = input("Enter the first name: ")
age1 = int(input(f"Enter the age for {name1}: "))

name2 = input("Enter the second name: ")
age2 = int(input(f"Enter the age for {name2}: "))

# original data
print("\n--- Original Data ---")
print(f"{name1} is {age1} years old.")
print(f"{name2} is {age2} years old.")

# swap ages
temp_age = age1
age1 = age2
age2 = temp_age

# last part
print("\n--- Swapped Data ---")
print(f"{name1} is now {age1} years old.")
print(f"{name2} is now {age2} years old.")
```
