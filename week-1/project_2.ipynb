```python
# input
p = float(input("Enter the principal amount (P): "))
r = float(input("Enter the interest rate as a percentage (e.g., 5 for 5%): "))
t = float(input("Enter the time in years (T): "))

# formula
amount = p * (1 + (r / 100) * t)

# result
print(f"The total amount after {t} years is: ${amount:.2f}")
```
