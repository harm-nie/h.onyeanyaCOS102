```python

pmt = float(input("Enter the periodic payment amount (PMT): "))
r = float(input("Enter the annual interest rate as a decimal (e.g., 0.05 for 5%): "))
n = int(input("Enter the number of compoundings per year (n): "))
t = float(input("Enter the number of years (t): "))


periodic_rate = r / n
total_periods = n * t



numerator = ((1 + periodic_rate) ** total_periods) - 1
amount = pmt * (numerator / periodic_rate)


print(f"The future value of the annuity is: ${amount:.2f}")
```
